### Задание 2

```python
num_horse = horse_ind.loc[:, ['rectal temperature', 'pulse', 'total protein']]
num_horse.head()
```

#### rectal temperature

```python
q1 = num_horse['rectal temperature'].quantile(0.25)
q3 = num_horse['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = num_horse[num_horse['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)]
blowout = pd.concat([num_horse, remove_outliers]).drop_duplicates(keep=False).sort_values('rectal temperature')
blowout.loc[~pd.isnull(blowout['rectal temperature']), :]
```

- зная, что средняя температура 37,5, то текущие выбросы считаю нормальными. При пульсе 140, температура 35.4 - не знаю что там внутри животного происходит... Возможно оно в состоянии шока. В общем оставляю как есть. 

#### pulse

```python
q1 = num_horse['pulse'].quantile(0.25)
q3 = num_horse['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = num_horse[num_horse['pulse'].between(lower_bound, upper_bound, inclusive=True)]
blowout = pd.concat([num_horse, remove_outliers]).drop_duplicates(keep=False).sort_values('pulse')
blowout.loc[~pd.isnull(blowout['pulse']), :]
```

- С выбросами пульса не все так однозначно. При среднем 72, и даже после 3 квартиля значения 88. в общем я бы оценил далее адекватность этих цифр в связке с другими показателями. Тут бы корреляцию провести с другими, но мы еще не умеем :) 

#### total protein

```python
q1 = num_horse['total protein'].quantile(0.25)
q3 = num_horse['total protein'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = num_horse[num_horse['total protein'].between(lower_bound, upper_bound, inclusive=True)]
blowout = pd.concat([num_horse, remove_outliers]).drop_duplicates(keep=False).sort_values('total protein')
blowout.loc[~pd.isnull(blowout['total protein']), :]
```

Далее собираюсь заполнять пустоты в зависимости от заполнения других величин.

```python
def change_values(column):
    list_for_check = list(horse_ind.columns)
    list_for_check.remove(column)
    
    while len(list_for_check) != 0:
    
        horse_ind[column].fillna(
        horse_ind.groupby(list_for_check)[column].transform('median'), inplace=True
            )
        
        if horse_ind[column].isnull().mean() == 0:
            break
        
        list_for_check.pop()
        if len(list_for_check) == 0:
            horse_ind[column].fillna(horse_ind[column].mode()[0], inplace=True)
```

```python
for column in horse_ind.columns:
    
    if horse_ind[column].isnull().mean() == 0:
        print(column)
        continue
    else:
        change_values(column)
```

```python
for col in horse_ind.columns:
    pct_missing = horse_ind[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```
