```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
df = pd.read_excel('hw_visual.xlsx')
```

```python
df.info()
```

```python
# ave_talk_time - среднее время разговора
# acw - среднее время постобработки
# aht - средрее время обработки звонка
df.columns = ['operator', 'group', 'value_calls', 'ave_talk_time', 'acw', 'aht']
```

```python
df.head()
```

```python
for col in ['aht','acw','ave_talk_time']:
    df[col] = pd.to_datetime(df[col], format = '%H:%M:%S')
    df[col] = df[col].dt.hour*60*60 + df[col].dt.minute*60 + df[col].dt.second
    df[col] = df[col].fillna(0).astype(np.int64)
```

```python
df.loc[df.group.str.contains('Группа 5'), 'department'] = 'Дист. отдел №1'
df.loc[df.group.str.contains('Группа 6'), 'department'] = 'Дист. отдел №2'
df.department.fillna('Офис', inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.loc[df.value_calls > 60,:] # очищаю от тех, кто принял звонков меньше чем за 1 день работы
```

```python
df.describe()
```

```python
df = df.loc[df.value_calls <= 1000,:] # очищаю от тех, кто принял звонков больше 1000, так как это скорее всязано с переработками
```

```python
sns.regplot(data=df, x='aht', y='acw', color='blue')
```

#### Вывод 3
Да, чем выше ACW, тем выше AHT

Далее попробую применить тепловую диаграмму на столь ничтожно малом количестве данных =) 

```python
df['acw_round'] = df.acw.apply(lambda x: (x // 100)*100)
```

```python
df['aht_round'] = df.aht.apply(lambda x: (x // 100)*100)
```

```python
group_aht = df.pivot_table(values='aht_round', index='group', columns='acw_round')
sns.heatmap(group_aht)
```

```python
table = df.pivot_table(values='value_calls', index='group', columns='department')
sns.heatmap(table)
```

```python
table = df.pivot_table(values='value_calls', index='aht_round', columns='acw_round')
sns.heatmap(table, cmap= 'PuBuGn')
```

#### Вывод 4
Принцип работы понятен. Но к моему датафрейму видимо не применим, чтобы было прям интересно :)
